```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/ex-3-1.csv')
df
```

```python
norm_dict = {
    'ID': df.ID,
    'DIST_NORM': (1 - np.e ** (1 - df.DISTANCE/df.DISTANCE.min())),
    'STOP_NORM': (1 - np.e ** (1 - df.STOP_COUNT/df.STOP_COUNT.min())),
    'COST_NORM': (1 - np.e ** (1 - df.COST/df.COST.min()))
}
norm_df = pd.DataFrame(norm_dict)
norm_df
```

```python
df = df.join(norm_df, how='inner', lsuffix='_left', rsuffix='_right').drop(['ID_left', 'ID_right'], axis=1)
df['NORM_SUM'] = df.DIST_NORM + df.STOP_NORM + df.COST_NORM
```

```python
df.sort_values(by='NORM_SUM')
```
