```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/restoraunts.csv')
df
```

```python
# exponential normalization
# y(x) = 1 - exp(1 - x/min(x))

df['gnorm'] = 1 - np.e**(1 - df['GUESTS']/df['GUESTS'].min())
df['inorm'] = 1 - np.e**(1 - df['INCOME']/df['INCOME'].min())
df['snorm'] = 1 - np.e**(1 - df['STARS']/df['STARS'].min())

df['NSUM'] = df['gnorm'] + df['inorm'] + df['snorm']
df.sort_values(by='NSUM', ascending=False).head(3)
```
